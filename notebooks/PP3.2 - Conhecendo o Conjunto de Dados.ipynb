```python
def obter_medidas(atributo):

    #dispersão desvio padrão absoluto, variancia, desvio padrão
    desvio_abs = dataset[atributo].mad()
    var = dataset[atributo].var()
    desvio_padrao = dataset[atributo].std()

    text = "desvio padrão abs:" + str(desvio_abs)+"\n variância:"+ str(var)+" \n desvio padrão: "+str(desvio_padrao)

    return text
```

```python
def gerar_grafico(atributo, medida):
        #medidas de tendencia central media, mediana, moda
    media = dataset[atributo].mean()
    mediana = dataset[atributo].median()
    moda = dataset[atributo].mode()

   
    ax = sns.distplot(dataset[atributo], kde= False)

    ax.figure.set_size_inches(12, 6)
    ax.set_title('Distribuição de '+str(atributo), fontsize = 18)
    ax.set_xlabel(medida, fontsize = 14)
    ax.set_ylabel('Quantidade', fontsize = 14)

    #anotações de tendencia central
    ax.annotate('média', xy=(media, 0), xytext=(media, 2000),arrowprops=dict(facecolor = 'g', arrowstyle="simple",connectionstyle="arc3",))
    ax.annotate('moda', xy=(moda, 0), xytext=(moda, 3000),arrowprops=dict(facecolor = 'g', arrowstyle="simple",connectionstyle="arc3",))
    ax.annotate('mediana', xy=(mediana, 0), xytext=(mediana, 4000),arrowprops=dict(facecolor = 'g', arrowstyle="simple",connectionstyle="arc3",))
    
    at = AnchoredText(obter_medidas(atributo), 
                        prop= dict(size = 12), frameon = True,
                        loc = 'upper left',)
    at.patch.set_boxstyle("round , pad = 0. , rounding_size = 0.2")
    ax.add_artist(at)
    ax
```

```python
wilderness_area = dataset.filter(regex='Wilderness_Area')
wilderness_area['Wilderness_Area'] = wilderness_area.idxmax(1)
wilderness_area['Wilderness_Area'] = wilderness_area['Wilderness_Area'].astype('category')
wilderness_area['Wilderness_Area'] = wilderness_area['Wilderness_Area'].cat.codes

ax = sns.distplot(wilderness_area['Wilderness_Area'] ,kde=False)

ax.figure.set_size_inches(12, 6)
ax.set_title('Distribuição da Area Selvagem', fontsize = 18)
ax.set_xlabel('Tipos Area Selvagem', fontsize = 14)
ax.set_xlim([0, 4])
ax.set_xticks([0, 1 ,2 ,3])
ax.set_ylabel('Quantidade', fontsize= 14)
```

```python
#proximo tratar atributos qualitativos
soil_type_column = dataset.filter(regex='Soil_Type')
soil_type_column['Soil_Type'] = soil_type_column.idxmax(1)
soil_type_column['Soil_Type'] = soil_type_column['Soil_Type'].astype('category')

# a = soil_type_column.value_counts()
soil_type_column['Soil_Type_Number'] = soil_type_column['Soil_Type'].cat.codes

ax = sns.distplot(soil_type_column['Soil_Type_Number'], kde= False)

ax.figure.set_size_inches(12, 6)
ax.set_title('Distribuição de Tipos de Solo', fontsize = 18)
ax.set_xlabel('Tipos de Solo', fontsize = 14)
ax.set_xlim([0, 40])
ax.set_xticks([i for i in range(0, 40)])
ax.set_ylabel('Quantidade', fontsize= 14)
```
